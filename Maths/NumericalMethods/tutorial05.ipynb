**Question 2: Multivariate Integration** $\\$

(a) Using `nquad` from `scipy` package, integrate $f(x,y,z) = (\frac{1}{\pi^{3/2}})*e^(-x^2 - y^2 - z^2)$ over $x=[-3,3], y=[-3,3]$ and $z=[-3,3]$. $\\$

(b) Compute the trunaction error.

```python
def func(x,y,z):
    ans = 1/pow(np.pi,3/2)
    return ans * pow(np.e,-x**2-y**2-z**2)
    
```

```python
xrange = [-3,3]
yrange = [-3,3]
zrange = [-3,3]

integral = nquad(func,[xrange,yrange,zrange] )[0]

print(f"Answer of the given integral is {integral}")
```
