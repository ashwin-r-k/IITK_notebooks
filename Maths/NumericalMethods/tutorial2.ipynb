**1. Riemanns Integral**

Riemann Integral approximation is given by, 

$\int_{a}^{b} f(x) dx \approx \sum_{i=0}^{n-1} h f(x_i)$, 

where $h=\frac{b-a}{n}$. $x_i$ is the function at $i^{th}$ point. $\\$
Consider the function $f(x) = sin(x)$ and $n=5$ on interval [$0,\pi$], and compute the Riemman integral for $\\$

(a) left endpoint

(b) right endpoint

(c) middle point

```python
def f(x):
    return np.sin(x)

def riemannsI(f,a,b,n,pos):
    
    h = (b-a)/n
    
    x = np.linspace(a,b,n+1)
    print(x)

    x[:] = x[:] + pos # for left 0 , mid h/2, right h
    
    sum = 0
    for i in range(n):
        sum =sum + h * f(x[i])
        #print(x[i])
    return sum



```

**3. Trapezoid Rule** $\\$

$\int_a^b f(x) dx \approx \sum_{i=0}^{n-1} h \frac{f(x_i) + f(x_{i+1})}{2}$ $\\$

Consider $f(x)  = sin(x)$ on [0,$\pi$], and $\\$

(a) compute the integral from trapezoid rule for n=5,50,100,1000,10000.$\\$

(b) Plot the trucntation error with n. $\\$

(c) Use scipy to compute the difference, and plot it with n. 


```python
def f_sin(x):
    return np.sin(x)

def trapezoidI(f,a,b,n):
    
    h = (b-a)/n
    
    x = np.linspace(a,b,n+1)
    #print(x)
    
    sum = 0
    for i in range(n):
        sum = sum + h * (f(x[i])+f(x[i+1]))/2
        #print(x[i])
    return sum

    
```

```python
a = 0
b = np.pi
n = [5,50,100,1000,10000]
integral = []
T_integral = []
print(scipy.integrate.quad(f_sin,a,b)[0])
for i in n:
    integral.append(trapezoidI(f_sin,a,b,i))
    T_integral.append(scipy.integrate.quad(f_sin,a,b)[0])
    print("For n = {0} Integral = {1}".format(i,integral[-1]))

    
plt.plot(n, np.array(T_integral[:]) - np.array(integral[:]), "*--b",label = "Error")
plt.xlabel("n number of intervals")
plt.ylabel(" Error")
plt.title("Errors in sin integral wrt number of intervals")
plt.legend()
plt.show()
```
