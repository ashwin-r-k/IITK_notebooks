```python
# Answer 1a: Fixed Point (Iterative)


def fixedp(given_function, x0, min_error=0.001, max_iteration=3) -> Tuple[float, List]:
    """ Fixed point algorithm """
    #start_time = time.time_ns()  
    i = 0
    error = 1
    xp = []
    x = None
    errors = []
    while error > min_error and i < max_iteration:
        x = given_function(x0)
        error = abs(x0 - x)
        x0 = x
        xp.append(x0)
        i += 1
        errors.append(error)
    #print((time.time_ns() - start_time)*1.e9, "nano seconds")
    #print(xp)
    return x, xp,errors

```

```python
def func(x):
    print(x)
    return 1-pow(x-1,2)

def g(x):
    return func(x) + x
    
a = 2
b = 3
n = 100

xline = np.linspace(a,b,n)
roots = [0,2]

plt.plot(xline,func(xline),label="function")
plt.scatter(roots,np.zeros(len(roots)),color = "k",label = "roots")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y = 0,ls = '--',color = "r")
plt.legend()
plt.show()
```

```python
# Answer 1b: Bisection Method (Covergent)
# Prints root of func(x)
# with error of EPSILON

def bisection(func,I):
    a = I[0]
    b = I[1]
    if (func(a) * func(b) >= 0):
        print("You have not assumed right a and b\n")
        return
    c = a
    while ((b-a) >= 0.01):
        c = (a+b)/2
        if (func(c) == 0.0):
            break
        if (func(c)*func(a) < 0):
            b = c
        else:
            a = c
    print("The value of root is : ","%.6f"%c)
    return c

I1 = [-3,0.5]
I2 = [0.5,3] 

print(bisection(func,I1))

print(bisection(func,I1))
```

```python
# Answer 1c: Regular Falsi Method. (Convergent)

# Prints root of func(x) in interval [a, b]
def regulaFalsi( func,I,MAX_ITER=10):
    a = I[0]
    b = I[1]
    if func(a) * func(b) >= 0:
        print("You have not assumed right a and b")
        return -1
    
    c = a # Initialize result
    
    for i in range(MAX_ITER):
        
        # Find the point that touches x axis
        c = (a * func(b) - b * func(a))/ (func(b) - func(a))
        
        # Check if the above found point is root
        if func(c) == 0:
            break
        
        # Decide the side to repeat the steps
        elif func(c) * func(a) < 0:
            b = c
        else:
            a = c
    print("The value of root is : " , '%.4f' %c)
    return c

Interval = [-3,0.5]
regulaFalsi( func,Interval,10)

Interval = [0.5,3]
regulaFalsi( func,Interval,10)
```

```python
# Answer 1d: Secant method (Iterative)

def secant(f,x1, x2, E):
	n = 0; xm = 0; x0 = 0; c = 0; 
	if (f(x1) * f(x2) < 0):
		while True: 
			
			# calculate the intermediate value 
			x0 = ((x1 * f(x2) - x2 * f(x1)) /
							(f(x2) - f(x1))); 

			# check if x0 is root of 
			# equation or not 
			c = f(x1) * f(x0); 

			# update the value of interval 
			x1 = x2; 
			x2 = x0; 

			# update number of iteration 
			n += 1; 

			# if x0 is the root of equation 
			# then break the loop 
			if (c == 0): 
				break; 
			xm = ((x1 * f(x2) - x2 * f(x1)) /
							(f(x2) - f(x1)));
			
			if(abs(xm - x0) < E):
				break;
		
		print("Root of the given equation =", 
							round(x0, 6)); 
		print("No. of iterations = ", n); 
		
	else:
		print("Can not find a root in ",
				"the given interval"); 
    

tollarance = 0.001
Interval = [0.5,3]

secant(func,Interval[0],Interval[1],tollarance)


```
