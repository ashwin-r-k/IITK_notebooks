## Q1 Compute following Fourier transforms of function f (x) using three versions (systems).




```python
# My definations for FT
def FT_integral(x,func,s,version=1):
    x = np.array(x)
    integral = 0
    dx = np.diff(x).mean() 
    if version == 1:
        integral = np.sum( func(x) * np.exp( -2* np.pi * 1j * x * s) *dx )
    elif version == 2:
        integral = np.sum(func(x) * np.exp( -1* 1j * x * s) * dx )
    elif version == 3:
        integral = np.sum(func(x) * np.exp( -1* 1j * x * s) * dx )/np.sqrt(2*np.pi)
    return integral 


def FourierTransform(x, func, version=1, s_min=-10, s_max=10, N=1000):
    s = np.linspace(s_min, s_max, N)
    FT = np.zeros(N, dtype=complex)
    for i in range(N):
        FT[i] = FT_integral(x, func, s[i],version)
    return s, FT

def FT_numpy(x, func):
    N = len(x)
    dt = x[1] - x[0]  
    f = np.fft.fftfreq(N, dt)  
    f = np.fft.fftshift(f) 
    FT = np.fft.fft(func(x)) * dt 
    FT = np.fft.fftshift(FT) 
    return f, FT

def plot_4(x,func):
    plt.figure(figsize=(10, 10))

    s, FT1 = FourierTransform(x, func, version=1)
    s, FT2 = FourierTransform(x, func, version=2)
    s, FT3 = FourierTransform(x, func, version=3)
    s_min, s_max = -10, 10
    # Using numpy's FFT
    f, FT = FT_numpy(x, func)

    plt.subplot(2,2,1)
    plt.plot(f, FT.real)
    plt.title('Fourier Transformusing Numpy')
    plt.xlim(s_min, s_max)
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.subplot(2,2,2)
    plt.plot(s, FT1.real)
    plt.legend()
    plt.title('Fourier Transform using version 1')
    plt.xlim(s_min, s_max)
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.subplot(2,2,3)
    plt.plot(s, FT2.real)
    plt.title('Fourier Transform using version 2')
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.subplot(2,2,4)
    plt.plot(s, FT3.real)
    plt.title('Fourier Transform using version 3')
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()

def plot_4img(x,func):
    plt.figure(figsize=(10, 10))

    s, FT1 = FourierTransform(x, func, version=1)
    s, FT2 = FourierTransform(x, func, version=2)
    s, FT3 = FourierTransform(x, func, version=3)
    s_min, s_max = -10, 10
    # Using numpy's FFT
    f, FT = FT_numpy(x, func)

    plt.subplot(2,2,1)
    plt.plot(f, FT.imag)
    plt.title('Fourier Transformusing Numpy')
    plt.xlim(s_min, s_max)
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.subplot(2,2,2)
    plt.plot(s, FT1.imag)
    plt.legend()
    plt.title('Fourier Transform using version 1')
    plt.xlim(s_min, s_max)
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.subplot(2,2,3)
    plt.plot(s, FT2.imag)
    plt.title('Fourier Transform using version 2')
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.subplot(2,2,4)
    plt.plot(s, FT3.imag)
    plt.title('Fourier Transform using version 3')
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()
```

```python
def scipy_fft(x, func):

    N = len(x)                 
    dt = x[1] - x[0]       
    FT = sp.fft.fft(func) * dt  
    f = sp.fft.fftfreq(N, d=dt)   
    f = sp.fft.fftshift(f)
    FT = sp.fft.fftshift(FT)
    return f, FT

# inverse FFt
def scipy_ifft(x, func):
    N = len(x)                 
    dt = x[1] - x[0]       
    FT = sp.fft.ifft(func) * dt  
    f = sp.fft.fftfreq(N, d=dt)   
    f = sp.fft.fftshift(f)
    FT = sp.fft.ifftshift(FT)
    return f, FT
```

```python
import scipy.signal as signal


def convolution(x, F, G):
    N = len(x)
    dt = x[1] - x[0]
    conv = np.zeros(N, dtype=complex)
    for i in range(N):
        shifted_G = np.roll(G, i)  # Shift G by i
        conv[i] = np.sum(F * shifted_G) * dt  # Compute convolution
    return conv

def window(x,w):
    N = len(x)
    windowed_func = np.zeros(N)
    for i in range(N):
        if x[i] <= w:
            windowed_func[i] = 1
        else:
            windowed_func[i] = 0
    return windowed_func

def smoothed_func(x, F, w):
    G = window(x,w)
    return convolution(x, F, G)

def smoothed_func2(x, F, w):
    G = window(x,w)
    return convolution(x, F, G)


def convolution_using_FT(x, f, g):

    N = len(x)
    dx = x[1] - x[0]  # Step size

    # Compute Fourier transforms
    F = sp.fft.fft(f)
    G = sp.fft.fft(g)

    # Convolution in Fourier space (Multiplication)
    conv_FT = F * G

    # Compute inverse Fourier transform (normalize by dx)
    conv_result = sp.fft.ifft(conv_FT) * dx

    return np.real(conv_result)  # Taking real part (should be real)

```
